```python
dia = 'martes'
if dia == 'martes':
    print('Si, es dia martes')
else:
    print("No, es otro día de la semana")
```

```python
dia = 'lunes'
if dia == 'martes':
    print('Si, es dia lunes')
elif dia == 'martes':
    print('Si, es dia martes')
else:
    print('No, es cualquier otro dia de la semana')
```
