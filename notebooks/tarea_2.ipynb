```python
# Importamos librerías necesarias.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Primera función: def descargar_datos(ticker, start, end)

def descargar_datos_yfinance(tickers, start, end):
    df = yf.download(tickers, start=start, end=end)['Close']
    return df  

tickers = ["AAPL", "JPM", "TSLA"]
start = "2021-01-01"
end = "2024-01-30"

df = descargar_datos_yfinance(tickers, start, end)
print(df)
```
