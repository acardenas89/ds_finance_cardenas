```python
# Importamos librerías necesarias.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Primera función: def descargar_datos(ticker, start, end)

def descargar_datos_yfinance(tickers, start, end):
    df = yf.download(tickers, start=start, end=end)['Close']
    return df  

tickers = ["AAPL", "JPM", "TSLA"]
start = "2021-01-01"
end = "2024-01-30"

df = descargar_datos_yfinance(tickers, start, end)
print(df)
```

```python
# Segunda función: def calcular_retornos(df):

def calcular_retornos(df):
    retornos = df.pct_change().dropna()
    return retornos

retornos = calcular_retornos(df)
print(retornos)
```

```python
# Función palra calcular: tendencia central, dispersión, localización y forma de los retornos.

def resumen_estadistico_retornos(retornos):
    resumen = pd.DataFrame({
        'media': retornos.mean(),
        'volatilidad': retornos.std(),
        'asimetria': retornos.skew(),
        'minimo': retornos.min(),
        'maximo': retornos.max(),
        'kurtosis': retornos.kurt()
    })
    return pd.DataFrame(resumen)

resumen = resumen_estadistico_retornos(retornos)
print(resumen)

```
