```python
# Ejercicio 15 
# Volver a estudiar

def maximo(lista):
    mayor = lista[0]

    for numero in lista:
        if numero > mayor:
            mayor = numero

    return mayor

numeros = [10, 5, 30, 8, 20]

print(maximo(numeros))

```

```python
# Ejercicio 17

def calcular_total(precio, cantidad):
    total = precio * cantidad

    if cantidad > 3:
        total = total * 0.90

    return total


precio = float(input("Precio del producto: "))
cantidad = int(input("Cantidad de unidades: "))

print("Total a pagar:", calcular_total(precio, cantidad))
```
